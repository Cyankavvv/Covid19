```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure

import requests
```

```python
url = 'https://api.covid19api.com/summary'
```

```python
r = requests.get(url)
```

```python
json = r.json()
```

```python
df = pd.DataFrame(json['Countries'])
```

```python
df.tail(10)
```

```python
plt.bar(df['Country'][0:10], df['TotalConfirmed'][0:10])
plt.xticks(rotation = 75, size = 14)
plt.grid()

plt.yticks(size = 14)
```

```python
df[ df['Country'] == 'Venezuela (Bolivarian Republic)']
```

```python
paises = ['Argentina', 'Brazil', 'Bolivia', 'Chile', 'Colombia','Ecuador', 'Peru', 'Paraguay', 'Uruguay', 'Venezuela (Bolivarian Republic)']
```

```python
for pais in paises:
    print(pais)
```

```python
df2 = pd.DataFrame()
for pais in paises:
    display(df[ df['Country'] == pais])
    df2 = pd.concat([df2, df[ df['Country'] == pais]], axis = 0)
```

```python
figure(figsize=(6, 5), dpi=120)

plt.bar(df2['Country'], df2['TotalConfirmed'])
plt.xticks(rotation = 75, size = 14)
plt.grid()
plt.ylabel(' Cantidad de casos confirmados')
plt.title('Casos confirmados C19')
plt.yticks(size = 14)
```

```python
df2['Relación fallecidos / casos confirmados'] = df2['TotalDeaths'] / df2['TotalConfirmed']
```

```python
figure(figsize=(5, 4), dpi=120)

plt.bar(df2['Country'], df2['Relación fallecidos / casos confirmados'])
plt.xticks(rotation = 75, size = 14)
plt.grid()
#plt.ylabel('Cantidad')
plt.title('Relación fallecidos / casos confirmados')
plt.yticks(size = 14)
```

```python
figure(figsize=(6, 4), dpi=120)

plt.bar(df2['Country'], df2['TotalDeaths'])
plt.xticks(rotation = 75, size = 14)
plt.grid()
plt.ylabel('Cantidad de personas fallecidas')
plt.title('Personas fallecidas por C19')
plt.yticks(size = 14)
```
